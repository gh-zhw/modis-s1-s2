```python
import matplotlib.pyplot as plt
import numpy as np


def show_image(image):

    rgb = image[:3, :, :]
    rgb = np.transpose(rgb, (1, 2, 0))

    rgb = (rgb-np.min(rgb)) / (np.max(rgb) - np.min(rgb))

    plt.imshow(rgb)
    plt.show()

```

```python
image = np.load("D:\Code\MODIS_S1_S2\dataset\SatelliteImages\S2\S2_42.npy")
show_image(image)
```
